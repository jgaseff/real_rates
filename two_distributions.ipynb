```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
n = 1000
mu_port = 7
sigma_port = 5.6
mu_bench = 3
sigma_bench = 3.3
bench = np.random.normal(mu_bench,sigma_bench,n)
port = np.random.normal(mu_port,sigma_port,n)

df1 = pd.DataFrame({'return': [*port, *bench], 'type': [*np.repeat('port',n), *np.repeat('bench',n)]})
fig1 = px.histogram(df1, x = 'return', color = 'type')
fig1.show()
```
